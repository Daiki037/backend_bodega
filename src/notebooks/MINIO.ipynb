```python
from minio import Minio
from minio.error import S3Error
```

```python
# Configuración del cliente MinIO
client = Minio(
    "localhost:9000",  # URL del servidor MinIO
    access_key="carlos",  # Access Key
    secret_key="12345678",  # Secret Key
    secure=False  # Cambiar a True si usas HTTPS
)

# Nombre del bucket
bucket_name = "prueba-udenar"

try:
    # Crear el bucket si no existe
    if not client.bucket_exists(bucket_name):
        client.make_bucket(bucket_name)
        print(f"Bucket '{bucket_name}' creado exitosamente.")
    else:
        print(f"Bucket '{bucket_name}' ya existe.")
    
    # Ruta del archivo a subir
    archivo_local = "D:/Carlos/Udenar-Investigación/Tesis/Documentos csun/LABORATORIOS/LABORATORIOS/LAB JIMENEZ/Q04-23-12952760.pdf"  # Cambia esto a la ruta de tu archivo
    archivo_remoto = "LabJimenez/Prueba.pdf"  # Nombre con carpeta en el bucket

    # Subir el archivo al bucket
    client.fput_object(bucket_name, archivo_remoto, archivo_local)
    print(f"Archivo '{archivo_local}' subido exitosamente a '{archivo_remoto}'.")
except S3Error as e:
    print(f"Error: {e}")
```
